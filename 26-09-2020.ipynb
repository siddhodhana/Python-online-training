**phone number validation**
- starting with +91
- must have 10 digits except country code
- must begin with any one of these numbers 6,7,8,9

ex - 9874563210

```python
pattern="^[6-9][0-9]{9}$||^[+][9][1][6-9][0-9]{9}$"
n=int(input())
import re
if re.match(pattern,str(n)):
    print(True)
else:
    print(False)
```
